## Task 2: Data Cleaning

#### Firstly match sepeators replace -'' with '/'.

```python
data['order_date'] = data['order_date'].str.replace('-', '/')
```

#### a. Convert the date to datetime formate.

```python
data['Order Date'] = pd.to_datetime(data['order_date'],format= '%d/%m/%Y',errors = 'coerce')
data['Ship Date'] = pd.to_datetime(data['ship_date'],format= '%d/%m/%Y',errors = 'coerce')
```

```python
data[['Order Date','Ship Date']].info()
```

## Task: 3 - Feature Engineering

#### a. Seperate month moth & year in a new column.

```python
data['Order_Month']  = data['Order Date'].dt.month_name()
```

```python
data[['Order_Month','year']]
```

```python
data.head(2)
```

## Task: 4- Profit analysis.

#### a. Analyse that which category having much profit.

```python
data.groupby('category')['profit'].sum().round(2).sort_values(ascending=False)
```

### b. Plot the information

```python
data.groupby('category')['profit'].sum().sort_values(ascending=False).plot(kind='bar',color=['blue','orange','green'])
plt.title('Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.show()
```

## Task: 5-

#### a. Firstly remove ',' from string sales so that it would change in float then int.

```python
data['sales'] = data['sales'].astype(float)
```

```python
data['sales'] = data['sales'].astype(int)
```

```python
data['sales'] = data['sales'].str.replace(',','')
```

#### b. find the sales according to month.

```python
data.groupby('Order_Month')['sales'].sum().sort_values(ascending = False)
```

#### c. now plot the data into bar chart.

```python
data.groupby('Order_Month')['sales'].sum().sort_values(ascending = False).plot(kind='bar',color=['blue','orange','green'])
plt.ticklabel_format(style='plain',axis = 'y') # this code is for removing scientific notation.
plt.title('Sales by Months')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.show()
```
