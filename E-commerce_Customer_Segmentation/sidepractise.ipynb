```python
import pandas as pd
```

```python
import numpy as np
```

```python
data = pd.DataFrame({
    'Customer': ['a','b','c','a','c','d'],
    'Date': ['2024-01-20','2024-01-25','2024-01-31','2024-01-01','2024-01-12','2024-01-17'],
    'Amount': [100,5000,200,1000,600,700]
    })
data
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data
```

```python
data.info()
```

```python
rfm = data.groupby('Customer').agg({'Date': lambda x: (pd.to_datetime('2024-02-01') - x.max()).days, 'Customer': 'count', 'Amount': 'sum'}) 
```

```python
rfm.columns = ['Recency','Frequency','Monetary']
rfm
```

```python
rfm['R_score'] = pd.qcut(rfm['Recency'],q=3,labels = [3,2,1])
rfm['F_score'] = pd.qcut(rfm['Frequency'].rank(method = 'first'),q=3,labels = [1,2,3])
rfm['M_score'] = pd.qcut(rfm['Monetary'],q=3,labels = [1,2,3])
```

```python
rfm
```

```python
rfm['RFM_Group'] = rfm['R_score'].astype(str) + rfm['F_score'].astype(str) + rfm['M_score'].astype(str)
rfm
```
