### a. filter the gdp data and bring only last year data.

```python
gdpdata = gdpdata[gdpdata['year'] == gdpdata['year'].max()]
gdpdata.head()
```

## Task 1:

#### a. Rename the column to country.

```python
gdpdata.rename(columns={'Country name': 'Country'}, inplace=True)
```

#### b. rename the column to country.

```python
coviddata.rename(columns={'Country/Region': 'Country'}, inplace=True)
```

## Task 2:

#### a. Merge both dataset.

```python
mergedata = pd.merge(gdpdata,coviddata, on= 'Country', how= 'inner')
```

```python
mergedata.info()
```

## Task: 3

#### a. Find the death rate.

```python
mergedata['Death_Rate'] =mergedata['Deaths'] / mergedata['Confirmed'] *100
```

```python
mergedata['Death_Rate']
```

#### b. Find the recovery rate.

```python
mergedata['Recovered_Rate'] =mergedata['Recovered'] / mergedata['Confirmed'] *100
```

```python
mergedata['Recovered_Rate']
```

```python
mergedata.head()
```

#### c. Find the country having much death rate.

```python
mergedata.sort_values(by='Death_Rate',ascending = False).head()
```

#### d. Find the country have highest gdp.

```python
mergedata.sort_values(by='Log GDP per capita',ascending = False).head()
```

#### e. Find the country having fast recovery rate.

```python
mergedata.sort_values(by='Recovered_Rate',ascending = False).head()
```

#### f. Finding that there is any impact of gdp on health.

```python
sns.scatterplot(x = 'Log GDP per capita', y = 'Death_Rate',data = mergedata)
plt.title('DEATH RATE CALCULATION OF COUNTRY HAVING HIGH GDP V/S LOW GDP')
plt.show()
```

```python
# this scatter plot tells that country having much gdp having much death rate except some, so the conclusion is that there are no impact of high gdp on health.
```
